### Hypothesis

Hypothesis:
If patient in the same demographic region, for a particular procedure/allregy is having very high cost of treatement, we consider it as a special case and there's high probability that the patient could be non adherent

```python
df_hypoth_1 = df_inter4[['Id_pat','Id_enc','STATE_pat', 'CITY_pat', 'ENCOUNTERCLASS_enc','ENCOUNTER_alg','DESCRIPTION_alg','TOTALCOST_med']]
df_hypoth_1.head()
```

```python
df_hypoth_1 = df_hypoth_1[~df_hypoth_1.ENCOUNTER_alg.isna()]
df_hypoth_1.head()
```

```python
df_hypoth_1[df_hypoth_1.Id_pat=='0288abb6-633c-40c3-ba0c-66c7d957727e']
```

```python
df_hypoth_1 = df_hypoth_1.groupby(['Id_pat','STATE_pat','ENCOUNTERCLASS_enc'])['TOTALCOST_med'].median().reset_index()
df_hypoth_1 = df_hypoth_1.dropna()
df_hypoth_1
```

```python
df_hypoth_1['percentile_cost'] = df_hypoth_1.groupby(['STATE_pat','ENCOUNTERCLASS_enc'])['TOTALCOST_med'].rank(pct=True, ascending=True)
df_hypoth_1['percentile_cost'].head()
```

```python
df_hypoth_1[df_hypoth_1.percentile_cost>0.95]
```

```python
list_high_cost_alg = df_hypoth_1[df_hypoth_1.percentile_cost>0.95]['Id_pat']
```
